```python
df_comments = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/news-edits/study2-reddit/bidencomments.csv')
df_comments.drop(df_comments.columns[df_comments.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df_comments
```

```python
def remove_urls(text):
    prog = re.compile(r'http\S+')
    text_processed =  prog.sub('<LINK>', str(text))
    return text_processed
def remove_hashtags(text):
    text = re.sub(r"#", '<HASHTAG>', str(text), flags=re.MULTILINE)
    return text

#Emoji regex has to be improved and more patterns need to be added
def de_emojify(text):
    regex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F" 
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
                           "]+", flags = re.UNICODE)
    return regex_pattern.sub('<EMOJI>', text)

def find_user_mentions(text):
    word=re.findall(r'(?<=@)\w+',text)
    return " ".join(word)

def user_mentions(text):
    text = re.sub(r'(?<=@)\w+', '<MENTION>', str(text), flags=re.MULTILINE)
    word=re.findall(r'(?<=@)\w+',text)
    return text
```

```python
df_comments['comment'] = df_comments['comment'].apply(remove_urls)
#df_comments['comment'] = df_comments['comment'].apply(remove_hashtags)
df_comments['comment'] = df_comments['comment'].apply(de_emojify)
df_comments['user_mentions']=df_comments['comment'].apply(lambda x: find_user_mentions(x))
df_comments['user_mentions']=df_comments['comment'].apply(user_mentions)
```
